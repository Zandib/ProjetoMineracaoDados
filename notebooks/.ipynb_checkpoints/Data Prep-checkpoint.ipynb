#### O intuito é realizar um data prep para suavizar o ruido nos audios, aumentando a similaridade entre eles

<p style="padding-left:15px">*<i>Todo o data prep realizado abaixo sera feito em cima da base de treino e replicado para a base de validação</i><p>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
from PIL import Image
import pathlib
import csv
from os import listdir
from os.path import isfile, join

from tqdm import tqdm
from tqdm import tqdm_notebook

import librosa

from scipy import signal

import warnings
warnings.filterwarnings('ignore')
```

### Visualização de alguns exemplos

```python
t_path = '../treinamento/'
letra_analisada = 'b'
t_path+=letra_analisada+'/'
onlyfiles = [f for f in listdir(t_path) if isfile(join(t_path, f))]

np.random.shuffle(onlyfiles)    
```

```python
number_of_test = 100
#fig, axs = plt.subplots(number_of_test)
dataframes = {}

for i in range(0,number_of_test):
    data,fs = librosa.load(t_path+onlyfiles[i])
    #Convertendo o dado para uma serie
    d_s = pd.Series(data)
    #Criando data_frame
    d_f = pd.DataFrame(data,columns=['dado'])
    #Normalizando d_s
    d_f['abs'] = d_s.abs()
    minM = d_f['abs'].min()
    maxM = d_f['abs'].max()
    d_f['norm'] = (d_f['abs'].apply(lambda x: (x-minM)/(maxM-minM)))*(d_f['abs']/d_f['dado'])
    dataframes[i] = d_f
    
#Shifting dataFrame vamos usar o dataframe 1 para igualar aos outros
#axs[0].plot(dataframes[0]['norm'])
new_dfs = {}
new_dfs[0] = dataframes[0]
for i in range(1,number_of_test):
    d0 = dataframes[0]
    target = dataframes[i]
    
    dx = np.mean(np.diff(d0.index.values))
    shift = (np.argmax(signal.correlate(d0['norm'],target['norm'])) - len(target['norm']))*dx
    #print(shift)
    #axs[i].set_ylim(-1,1)
    #axs[i].figure.set_figwidth(15)
    #axs[i].figure.set_figheight(17)
    #axs[i].plot(target.index+shift,target['norm'])
    #axs[i].plot(target.index,target['norm'])
    new_dfs[i] = target.copy()
    new_dfs[i].index=target.index+shift
    

#     axs[i].figure.set_figwidth(15)
#     axs[i].figure.set_figheight(7)
#     axs[i].set_ylim(-1,1)
#     axs[i].plot(d_f['norm'])
```

```python
fig,axs=plt.subplots(2)
plt.subplots_adjust(hspace = 0.4)
for i in new_dfs.keys():
    axs[0].plot(dataframes[i]['dado'])
    axs[0].figure.set_figheight(7)
    axs[0].figure.set_figwidth(15)
    axs[0].set_title('Dado original')
    axs[1].plot(new_dfs[i]['dado'])
    axs[1].set_title('Dado Alinhado')
fig.savefig('../imagens/amostragem_dados')
```
